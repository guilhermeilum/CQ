```python
import pygame
import sys
import random
import math
from itertools import combinations

pygame.init()

WINDOW_WIDTH, WINDOW_HEIGHT = 800, 600
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Partículas")

WHITE = (255, 255, 255)


class Particle:
    def __init__(self, x, y, color):
        self.x = random.uniform(0, x)
        self.y = random.uniform(0, y)
        self.radius = 5
        self.speed_x = random.uniform(-1, 1)
        self.speed_y = random.uniform(-1, 1)
        self.color = color

    def move(self):
        self.x += self.speed_x
        self.y += self.speed_y

        self.x = max(0, min(self.x, WINDOW_WIDTH))
        self.y = max(0, min(self.y, WINDOW_HEIGHT))

    def draw(self):
        pygame.draw.circle(window, self.color, (int(self.x), int(self.y)), 5)

    def check_collision_with_boundaries(self):
        if self.x - self.radius <= 0 or self.x + self.radius >= WINDOW_WIDTH:
            self.speed_x *= -1

        if self.y - self.radius <= 0 or self.y + self.radius >= WINDOW_HEIGHT:
            self.speed_y *= -1


def check_collision_with_particle(particulas):
    lista_nova = particulas.copy()
    posição = []
    for index1, index2 in combinations(range(len(particulas)), 2):
        particula1 = particulas[index1]
        particula2 = particulas[index2]

        distancia = check_collision(particula1, particula2)
        if distancia <= particula1.radius + particula2.radius:
            chance = random.random()
            if chance > 0.8:
                particula1.speed_x *= -1
                particula1.speed_y *= -1

                particula2.speed_x *= -1
                particula2.speed_y *= -1
            else:
                if not (
                    particulas[index1].color == (0, 255, 0)
                    or particulas[index2].color
                    == (
                        0,
                        255,
                        0,
                    )
                ):
                    posição_x = (particula1.x + particula2.x) / 2

                    posição_y = (particula1.y + particula2.y) / 2

                    lista_nova.pop(index1)
                    lista_nova.pop(index2)
                    posição.append((posição_x, posição_y))

        for posição_x, posição_y in posição:
            particula = Particle(posição_x, posição_y, (0, 255, 0))
            lista_nova.append(particula)
        particulas = lista_nova

        # elif distancia <= 20:
        #     particula1.speed_x += particula1.speed_x*distancia/1000
        #     particula1.speed_y += particula1.speed_y*distancia/1000

        #     particula2.speed_x += particula2.speed_x*distancia/1000
        #     particula2.speed_y += particula2.speed_y*distancia/1000


def check_collision(particula1, particula2):
    # Calcula a distância entre os centros das partículas usando o teorema de Pitágoras
    distance = math.sqrt(
        (particula1.x - particula2.x) ** 2 + (particula1.y - particula2.y) ** 2
    )

    # Verifica se a distância é menor ou igual à soma dos raios das partículas
    return distance


def main():
    particulas = [
        Particle(WINDOW_WIDTH, WINDOW_HEIGHT, (255, 0, 0)) for _ in range(100)
    ]

    clock = pygame.time.Clock()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

        window.fill(WHITE)

        check_collision_with_particle(particulas)

        for particle in particulas:
            particle.move()
            particle.check_collision_with_boundaries()
            particle.draw()

        pygame.display.flip()
        clock.tick(60)


if __name__ == "__main__":
    main()
```
